```python
import batman
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(2454966.6983 - 1, 2454966.6983 + 1, 1000) 

params1 = batman.TransitParams()       # object to hold parameters for first planet
params1.t0 = 2454966.6983              # time of inferior conjunction
params1.per = 289.863876               # orbital period in days
params1.rp = 0.0275                    # planet radius (in units of stellar radii)
params1.a = 94.4                       # semi-major axis (in units of stellar radii)
params1.inc = 89.76                    # orbital inclination (in degrees)
params1.ecc = 0                        # eccentricity
params1.w = 90                         # longitude of periastron (in degrees)
params1.limb_dark = "quadratic"        # limb darkening model
params1.u = [0.630, 0.076]             # limb darkening coefficients

params2 = batman.TransitParams()       # object to hold parameters for first planet
params2.t0 = 2454966.6983              # time of inferior conjunction
params2.per = 289.863876               # orbital period in days
params2.rp = params1.rp/2                   # planet radius (in units of stellar radii)
params2.a = 94.4                       # semi-major axis (in units of stellar radii)
params2.inc = 89.76                    # orbital inclination (in degrees)
params2.ecc = 0                        # eccentricity
params2.w = 90                         # longitude of periastron (in degrees)
params2.limb_dark = "quadratic"        # limb darkening model
params2.u = [0.630, 0.076]             # limb darkening coefficients

m1 = batman.TransitModel(params1, time)
flux1 = m1.light_curve(params1)

m2 = batman.TransitModel(params2, time)
flux2 = m2.light_curve(params2)

fig = plt.figure(figsize=(10, 6))
plt.plot(time, flux1, color="teal", label="Kepler-22 b")
plt.plot(time, flux2, color="salmon", label="Second Planet")
plt.xlabel("Time from transit center (julian days)")
plt.ylabel("Relative flux")
plt.title("Transit Light Curves of Kepler-22 b and Second Planet")
plt.legend()
plt.show()
```
